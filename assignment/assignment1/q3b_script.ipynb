```python
# Author: Xinyu Ma
# Student ID: 28652703
import math
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
b = 0.04
N = 1000
mu = 0.03
beta = 0.9
n = 0.5
alpha = 0.1
def SEIR(P,t):
    s = P[0]
    e = P[1]
    i = P[2]
    r = P[3]
    dsdt = b - (beta*s*i) - mu*s
    dedt = (beta*s*i) - n*e - mu*e
    didt = n*e - mu*i - alpha*i
    drdt = alpha*i - mu*r
    return [dsdt, dedt, didt, drdt]
t = np.linspace(0,99,100)
P0 = np.array([0.9,0.001,0.01,0])
P = odeint(SEIR,P0,t)
plt.plot(t,P[:,0],label='Susceptible')
plt.plot(t,P[:,1],label='Exposed')
plt.plot(t,P[:,2],label='Infected')
plt.plot(t,P[:,3],label='Recovered')
plt.title("SEIR model with demography")
plt.legend()
plt.show()
```

```python
b = 0.04
N = 1000
mu = 0.03
beta = 0.9
n = 0.5
alpha = 0.1
p = 0.8
q = 0.7
r = 0.3
def SEIAR(P,t):
    s = P[0]
    e = P[1]
    i = P[2]
    a = P[3]
    r = P[4]
    dsdt = b - beta*s*i*p - q*s*a*(1-p) - mu*s
    dedt = beta*s*i*p + q*s*a*(1-p) - p*e - (1-p)*e - mu*e
    didt = p*e - alpha*i - mu*i
    dadt = (1-p)*e - r*a - mu*a
    drdt = alpha*i + r*a - mu*r
    return [dsdt, dedt, didt, dadt, drdt]
t = np.linspace(0,99,100)
P0 = np.array([0.9,0.001,0.01,0,0])
P = odeint(SEIAR,P0,t)
plt.plot(t,P[:,0],label='Susceptible')
plt.plot(t,P[:,1],label='Exposed')
plt.plot(t,P[:,2],label='Infected')
plt.plot(t,P[:,3],label='Asymptomatic')
plt.plot(t,P[:,4],label='Recovered')
plt.title("SEIAR model with demography")
plt.legend()
plt.show()
```

The proportion of infected people will decrease because the asymptomatic population is considered in our SEIAR model. The larger the p-value, the smaller the proportion of asymptomatic infections.
