```python
# Author: Xinyu Ma
# Student ID: 28652703
import math
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def x2y(x):
    return x/(x+2)
```

```python
E = sys.float_info.epsilon
```

```python
def my_log1p(x):
    log_approx = 0
    n = 1
    while n <= 1000000:
        coeff = (-1)**(n+1)
        num = x**n
        dem = n
        if abs((coeff*num)/dem) < 10**(-6):
            break
        log_approx += (coeff*num)/dem
        n += 1
    return log_approx,n-1,abs((log_approx-np.log1p(x))/np.log1p(x))
```

```python
def my_another_log1p(x):
    log_approx = 0
    n = 1
    y = x2y(x)
    while n <= 1000000:
        coeff = y**(2*n-1)
        num = 2
        dem = 2*n-1
        if abs((coeff*num)/dem) < 10**(-6):
            break
        log_approx += ((coeff*num)/dem)
        n += 1
    return log_approx,n-1,abs((log_approx-np.log1p(x))/np.log1p(x))
```

```python
print(my_another_log1p(1))
```

```python
x = np.linspace(-0.9,1,100)
y = []
y1 = []
for i in x:
    y.append(my_another_log1p(i)[0])
    y1.append(np.log1p(i))
plt.plot(x,y,label='another_log approx')
plt.plot(x,y1,label='built-in log')
plt.legend()
plt.show()
```

```python
x = np.linspace(-0.9,1,100)
y = []
y1 = []
y2 = []
for i in x:
    y.append(my_another_log1p(i)[2])
    y1.append(my_log1p(i)[2])
    y2.append(abs((np.log(1+i)-np.log1p(i))/np.log1p(i)))
plt.plot(x,y,label='another_log approx')
plt.plot(x,y1,label='my_log approx')
plt.plot(x,y2,label='actual log')
plt.title('relative error')
plt.legend()
plt.show()
```
